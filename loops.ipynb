```python

count=0
for x in range(100,1000):
    if(x%2==0 and x%3==0):
        print(x,end=', ')
        count+=1
print("\n------------------------------------------------------------------")
print("\nCount:",count)
```

```python
count=0
x=100
while(x<1000):
    if(x%2==0 and x%3==0):
        print(x,end=', ')
        count+=1
    x+=1
print("\n----------------------")
print("\nCount:",count)
```
